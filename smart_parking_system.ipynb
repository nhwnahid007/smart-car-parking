```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def _plot_series(series, series_name, series_index=0):
    """Plots a single series with a specific color from the palette."""
    palette = list(sns.palettes.mpl_palette('Dark2'))
    xs = series['Time']
    ys = series['Vehicle Count']
    plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

downsampled_df = df[df.index % 5 == 0]


fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')

for i, (series_name, series) in enumerate(downsampled_df.groupby('Status')):
    _plot_series(series, series_name, i)

fig.legend(title='Status', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Time')
plt.ylabel('Vehicle Count')
plt.title('Vehicle Count Over Time')
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
import matplotlib.pyplot as plt

gate_counts_data = {'Gate': ['Entry', 'Exit'],
                   'Entry': [50, 30],
                   'Exit': [40, 60]}

gate_counts_df = pd.DataFrame(gate_counts_data)

plt.figure(figsize=(8, 6))

# Entry in green
plt.bar(gate_counts_df['Gate'], gate_counts_df['Entry'], label='Entry', color='green')
# Exit in red
plt.plot(gate_counts_df['Gate'], gate_counts_df['Exit'], marker='o', linestyle='-', label='Exit', color='red')

plt.xlabel('Gate')
plt.ylabel('Count')
plt.title('Gate Counts by Status')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
